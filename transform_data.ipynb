```python
# Data baru
new_data = {'amount': 2123.00,
            'transaction_time': '2023-05-07 02:02:02',
            'location': 'City-30',
            'customer_age': 42,
            'card_type': 'MasterCard',
            'purchase_category': 'Gas Station'
            }

# Konversi data baru menjadi DataFrame
new_df = pd.DataFrame([new_data])

new_df.info()
```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])  # Mengubah kolom menjadi tipe data datetime64
# Mengubah format waktu
df['transaction_time'] = df['transaction_time'].dt.strftime('%H:%M:%S')
df['transaction_time'] = df['transaction_time'].str.replace(':', '')
df['transaction_time'] = df['transaction_time'].astype("int64")
```

```python
new_df['transaction_time'] = pd.to_datetime(new_df['transaction_time'])  # Mengubah kolom menjadi tipe data datetime64
# Mengubah format waktu
new_df['transaction_time'] = new_df['transaction_time'].dt.strftime('%H:%M:%S')
new_df['transaction_time'] = new_df['transaction_time'].str.replace(':', '')
new_df['transaction_time'] = new_df['transaction_time'].astype("int64")
```

```python
# Baca file txt
card_type_mapping_df = pd.read_csv('card_type.txt', sep='\t')
purchase_category_mapping_df = pd.read_csv('purchase_category.txt', sep='\t')
location_mapping_df = pd.read_csv('location.txt', sep='\t')

# Buat kamus pemetaan
card_type_mapping = dict(zip(card_type_mapping_df['card_type'], card_type_mapping_df['card_type_encode']))
purchase_category_mapping = dict(zip(purchase_category_mapping_df['purchase_category'], purchase_category_mapping_df['purchase_category_encode']))
location_mapping = dict(zip(location_mapping_df['location'], location_mapping_df['location_encode']))
```

```python
location_mapping_df
```

```python
# Lakukan pemetaan menggunakan metode map
new_df['card_type'] = new_df['card_type'].map(card_type_mapping)
new_df['purchase_category'] = new_df['purchase_category'].map(purchase_category_mapping)
new_df['location'] = new_df['location'].map(location_mapping)
```

```python
kolom = new_df.columns 
for i in range(len(kolom)) :
    if (new_df[kolom[i]].dtype == 'int64' or new_df[kolom[i]].dtype == 'float64'):
        if kolom[i] != "amount" :
            # Jika datanya interval atau ratio
            min_a = tes_df[kolom[i]].min()
            max_a = tes_df[kolom[i]].max()
            print(min_a)
            print(max_a)
            max_b = 1
            min_b = 0
            new_df[kolom[i]] = ((new_df[kolom[i]] - min_a) / (max_a - min_a)) * (max_b - min_b) + min_b

```
